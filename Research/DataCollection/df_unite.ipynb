```python
import pandas as pd
```

```python
# Read local data and concatenate it into one dataset
df_interfax = pd.read_parquet(r'interfax_parsed_data.parquet', index_col=0)

df_smart_labs = pd.read_parquet(r'smart_labs_parced_data.parquet')
df_smart_labs.rename(columns={'key_words': 'tags'}, inplace=True)

df_ria = pd.read_parquet(r'ria_parsed_data.parquet', index_col=0)

df_kommersant = pd.read_parquet(r'kommersant_parced_data.parquet')
df_kommersant.rename(columns={'key_words': 'tags'}, inplace=True)

# New df
df = pd.concat([df_interfax, df_smart_labs, df_ria, df_kommersant],
               axis=0,
               ignore_index=True)
# Save
df.to_parquet('Data/full_data_5years.parquet', index=False)
```
